<h4>Задание 4</h4>

```python
import json
import csv


with open('sales.json','r',encoding='utf-8') as f:
    data = json.load(f)    

with open('result.csv', "w", newline="") as file:
    columns = ['item', 'country', 'year', 'sales']
    writer = csv.DictWriter(file, delimiter = ';', fieldnames=columns)
    writer.writeheader()   
    
    for some_item in data:
        for country,y_s in some_item['sales_by_country'].items():
            for year in y_s:
                writer.writerow({'item':some_item['item'],'country':country,'year':year,'sales':y_s[year]})
```

В результате "pprint" почему-то решает, что надо в таком порядке отсортировать, если им не пользоваться то все по порядку идет

```python
import requests
import locale
import xml.etree.ElementTree as ET
from pprint import pprint
from copy import copy
from datetime import date,timedelta,datetime

def iter_datetime(start_date,end_date):
    while start_date < end_date:
        yield start_date.strftime("%d/%m/%Y")
        start_date = start_date + timedelta(days=1)

def dictify(r,root=True):
    if root:
        return {r.tag : dictify(r, False)}
    d=copy(r.attrib)
    if r.text:
        d["_text"]=r.text
    for x in r.findall("./*"):
        if x.tag not in d:
            d[x.tag]=[]
        d[x.tag].append(dictify(x,False))
    return d

api_str = 'http://www.cbr.ru/scripts/XML_daily.asp?date_req='

result_data = {}

generate_date = iter_datetime(datetime(2020,2,1),datetime(2020,7,2))
for day in generate_date:
    response = requests.get('http://www.cbr.ru/scripts/XML_daily.asp?date_req='+ day)
    data = response.text
    root = ET.fromstring(data)
    d = dictify(root)
    result_data.update({day:[]})
    for valute in d['ValCurs']['Valute']:
        if valute['ID'] == 'R01235':
            str_valute = valute['Value'][0]['_text']
            float_valute = float(str_valute.replace(',','.'))
            result_data[day].append({valute['Name'][0]['_text']:float_valute})
        if valute['ID'] == 'R01239':
            str_valute = valute['Value'][0]['_text']
            float_valute = float(str_valute.replace(',','.'))
            result_data[day].append({valute['Name'][0]['_text']:float_valute/10})
        if valute['ID'] == 'R01270':
            str_valute = valute['Value'][0]['_text']
            float_valute = float(str_valute.replace(',','.'))
            result_data[day].append({valute['Name'][0]['_text']:float_valute/100})
        if valute['ID'] == 'R01720':
            str_valute = valute['Value'][0]['_text']
            float_valute = float(str_valute.replace(',','.'))
            result_data[day].append({valute['Name'][0]['_text']:float_valute/10})
pprint(result_data)
        
    
    
```
